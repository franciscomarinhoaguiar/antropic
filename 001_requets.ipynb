```python
#Create an api client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)
    
    
```

```python
def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)
    
```

```python
# Make a request to the model
def chat(messages):
    message =  client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text


```

```python
# Make a starting list of messages
messages = []

# Add a user message
add_user_message(messages, "Hello, what is quantum computing? Answer in one sentence")

messages

# pass the lis of messages into çhat to get an answer
answer = chat(messages)

answer

# take the answer and add it as an assistante message into our list of messages
add_assistant_message(messages, answer)

messages

# add in the user's follow-up question
add_user_message(messages, "write another sentence")

messages

# call chat again to get the final answer
final_answer = chat(messages)

final_answer





```
